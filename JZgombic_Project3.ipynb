In the next step, we write the SQL Query to get the required information and we run that through **sqlite3**. 

The result that we get are the first 5 rows of data which have been retrieved using our query.

```python
QUERY = '''
        SELECT 
            LastName AS 'Last Name', 
            FirstName AS 'First Name',  
            Track.Name AS 'Track Name', 
            Album.Title AS 'Album Title'
         FROM 
             Customer
         INNER JOIN 
             Invoice
         ON 
             Customer.CustomerId = Invoice.CustomerId
         INNER JOIN 
             InvoiceLine
         ON 
             Invoice.InvoiceId = InvoiceLine.InvoiceId
         INNER JOIN 
             Track
         ON 
             InvoiceLine.TrackId = Track.TrackId
         INNER JOIN 
             Album
         ON 
             Track.AlbumId = Album.AlbumId
         ORDER BY 
             Customer.LastName, Customer.FirstName
         '''

con = sql.connect("Chinook_Sqlite.sqlite")

df = pd.read_sql_query(QUERY, con)

df.head(5)
```
